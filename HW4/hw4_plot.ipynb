```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path_1 = "/mnt/c/Users/HP/Documents/TTIC/31220_Unsupervised/HW4/geyser.txt"
```

```python
dur_list = []
inter_list = []
with open(file_path_1,'r') as f:
    for line in f:
        dur,inter = line.strip().split()
        dur_list.append(float(dur))
        inter_list.append(float(inter))
```

```python
geyser_np = np.array([dur_list,inter_list])
```

```python
t = np.linspace(0., 295, geyser_np.shape[1])
```

```python
# red dashes, blue squares and green triangles


# plt.figure(figsize=(20, 20))
plt.plot(t, geyser_np[0,:], 'r^',label='duration')
plt.xlabel('time step')
plt.ylabel('duration time')
plt.title(f'Geyser Duration vs. time step scatter')
plt.show()
```

```python
plt.figure(figsize=(20, 20))
plt.plot(t, geyser_np[0,:], 'r-',label='duration')
plt.plot(t, geyser_np[0,:], 'b^',label='duration')
plt.xlabel('time step')
plt.ylabel('duration time')
plt.title(f'Geyser Duration vs. time step',fontsize=50)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
plt.plot(t, geyser_np[1,:], 'r-',label='time till next')
plt.plot(t, geyser_np[1,:], 'b^',label='time till next')

plt.xlabel('time step')
plt.ylabel('time until next')
plt.title(f'time till next vs. time step',fontsize=50)
plt.show()
```
